## Q1

#### Question

Consider the applications used for managing courses and enrollments at Columbia, e.g. CourseWorks, Vergil. These applications use database management systems. Consider an alternate approach in which the university used a set of shared Google sheets to manage classes and enrollments. Faculty, administrators and students would edit the sheets to create classes, enroll in classes, etc. Identify five problems with the shared sheet approach and explain the features of a DBMS that solve the problem. Be succinct.

#### Answer

1. Data Integrity: Sheets allow typos and duplicate entries; a DBMS uses constraints and normalization to ensure data remains consistent and accurate.
2. Concurrency: Sheets fail when hundreds of students enroll simultaneously; a DBMS uses Transaction Management (ACID) to handle thousands of concurrent edits safely.
3. Security: Sheets offer limited "all-or-nothing" access; a DBMS provides Granular Access Control, ensuring students can only see their own private data.
4. Scalability: Sheets lag as data grows; a DBMS uses Indexing to maintain high-speed performance even with millions of enrollment records.
5. Data Modeling: Sheets are flat; a DBMS uses a Relational Model to easily map complex connections, such as Many-to-Many relationships between students and courses.

Briefly explain the concepts of _declarative_ and _procedural_ database manipulation languages. What is one benefit of declarative languages?

#### Answer

Declarative - stating what data is needed without the HOW. SQL is an example of such language.
Procedural - user must specify the exact steps, like a recipe, to get the data. this is the HOW. Popular programming languages are like so: Python, Java, etc.

The benefit of a declarative langauge is simplicity. The programmer doesn't need to care how to get the data, they just write in a short form what data they need.

### P9

#### Question

In this question, you use the ```db_book``` tables ```student, instructor``` and ```advisor```. You are recreating the $relational \ algebra$ query from question ```P3``` in SQL. Enter your SQL in the cell below. My answer, which you must match, is below. The order of the rows does not matter.

<img src="sql9-answer.jpg">

#### Answer

````python
%%sql
use db_book;

SELECT
    i.ID AS instructor_id,
    i.name AS instructor_name,
    'advises' AS advises,
    s.ID AS student_id,
    s.name AS student_name
FROM instructor i
JOIN advisor a ON i.ID = a.i_id
JOIN student s ON a.s_id = s.ID
WHERE i.dept_name = 'Comp. Sci.';
````
